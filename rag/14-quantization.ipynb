```python
import cv2
import matplotlib.pyplot as plt

# 
image_path = 'sample_data/src.jpeg'  # 
original_image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

#  Binary Quantization
_, binary_quantized_image = cv2.threshold(original_image, 128, 255, cv2.THRESH_BINARY)

# 
plt.figure(figsize=(10, 5))

# 
plt.subplot(1, 2, 1)
plt.imshow(original_image, cmap='gray')
plt.title('Original Image')
plt.axis('off')

#  Binary Quantization 
plt.subplot(1, 2, 2)
plt.imshow(binary_quantized_image, cmap='gray')
plt.title('Binary Quantized Image')
plt.axis('off')

plt.suptitle("Binary Quantization: Similarity Before and After Compression", fontsize=16)
plt.show()
```
